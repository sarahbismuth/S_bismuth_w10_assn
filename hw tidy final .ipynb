```python
import pandas as pd 
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/sarahbismuth/DATA-HW/master/DATA%20HW%20TEST.csv")
df
```

```python
#I use the melt function in order to transfrom the wide csv into a long
long_df=pd.melt(df, id_vars =['Unnamed: 0','Unnamed: 1'], value_vars =['Phoenix','San Francisco','San Diego','Seattle','Los angeles'], 
              var_name ='cityoforigin', value_name ='numberofflight') 
long_df
```

```python
#change the name of the column 
long_df.columns = ['airline', 'situation','cityoforigin','numberofflight']
```

```python
#for the airline  column all the Nan value will be replace by the previous data. 
long_df['airline'].fillna(method='ffill', inplace = True)
long_df
```

```python
#It will drop the Nan value 
df2 = long_df.dropna()
df2
```

```python
#this will give for the 2 airlines  the average number of flights  on time and delayed.
average_type = df2.groupby(['airline','situation']).numberofflight.mean()
average_type = average_type.reset_index()
average_type=average_type.sort_values('numberofflight',ascending=[0])
average_type.head()
```

From this we can see that the average delayed flight is higher in AM West than In Alaska

```python
#From long to wide 
df_wide = df2.reset_index().pivot_table(index = ['airline','situation'], columns = 'cityoforigin', values = 'numberofflight')
df_wide
```

I have decided to use the airline and the situation(on time or delayed) as the index, the city of origin ad the column and the number of flights as the value. 
After cleaning the original data as tidy data, by removing the Nan values and rename the column i think that this format is more convenient in this case than the long one, and that it is easier to understand the value in the wide format.
